```python
import pandas as pd
import numpy as np
from scipy.stats import entropy
from collections import defaultdict
import time
import math
```

```python
train2 = pd.read_csv("../data/train2.csv")
train2 = train2.drop(['Unnamed: 0'], axis=1)
train2['user_id'] = [i for i in range(1,len(train2)+1)]
train2.set_index('user_id',inplace=True)
train2.head()
#train2 = train2[]
```

```python
N_USER = len(train2)
#N_USER = 5
ENPY = defaultdict(lambda : defaultdict(float))
```

```python
for i in range(1,N_USER+1):
    #item to it-self has entropy 0
    ENPY[i][i] = 0.
    if i % 500 == 0:
        start_time = time.time()
    
    for j in range(i,N_USER+1):
        if j > i:
            Ri = train2.loc[i]
            Rj = train2.loc[j]
            Dij = (Ri-Rj).dropna(axis=0)

            if len(Dij) > 0:
                #if there exists common rated items
                ed = 0.
                for d in set(Dij):
                    Pd = float(Dij.tolist().count(d))/len(Dij)
                    if Pd > 0:
                        ed += - Pd * math.log(Pd, 2)
                ENPY[i][j] = ed
                ENPY[j][i] = ed
        
            else:
                #o.w.,assign entropy 999
                ENPY[i][j] = 999
                ENPY[j][i] = 999
    if i % 500 == 0:
        print "---Finish computing entropy score for",i,"users---"
        print "--- %s seconds ---" % (time.time() - start_time)
```

```python
df = pd.DataFrame.from_dict(ENPY) #convert nested dict to dataframe and save to csv
```

```python
train = pd.read_csv("../data/MS_train2.csv")
train = train.drop(['Unnamed: 0'], axis=1)
train.set_index('User',inplace=True)
train.fillna(0.0, inplace=True, axis=1)
train.head()
```

```python
N_USER = len(train)
ENPY1 = defaultdict(lambda : defaultdict(float))
USER_IDS = train.index
```

```python
for i in range(0,N_USER):
    #item to it-self has entropy 0
    ENPY1[USER_IDS[i]][USER_IDS[i]] = 0.
    if i % 500 == 0:
        start_time = time.time()
    
    for j in range(i,N_USER):
        if j > i:
            Ri = train.iloc[i]
            Rj = train.iloc[j]
            Dij = abs(Ri-Rj)
            
            if len(Dij) > 0:
                #if there exists common rated items
                ed = 0.
                for d in set(Dij):
                    Pd = float(Dij.tolist().count(d))/len(Dij)
                    if Pd > 0:
                        ed += - Pd * math.log(Pd, 2)
                ENPY1[USER_IDS[i]][USER_IDS[j]] = ed
                ENPY1[USER_IDS[j]][USER_IDS[i]] = ed
        
            else:
                #o.w.,assign entropy 999
                ENPY1[USER_IDS[i]][USER_IDS[j]] = 999
                ENPY1[USER_IDS[j]][USER_IDS[i]] = 999
    if i % 500 == 0:
        print "---Finish computing entropy score for",i,"users---"
        print "--- %s seconds ---" % (time.time() - start_time)
```

```python
df1 = pd.DataFrame.from_dict(ENPY1) #convert nested dict to dataframe and save to csv
df1.head()
```
